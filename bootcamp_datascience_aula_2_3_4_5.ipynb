```
import pandas as pd
dados = pd.read_csv("/content/drive/MyDrive/Estudo/Bootcamp Data Science/dados/A160324189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows=3, sep=';', skipfooter=12, thousands=".", decimal=",")
dados.head()
```

**Leitura do arquivo**
* Encoding: serve para interpretar certo acentuação
* skiprows: ignora linhas desnecessárias do ínicio do arquivo
* sep: separador do arquivo
* skipfooter: ignora linha desnecessárias do final do arquivo
* thousand: indica qual deve ser o separador de milhar
* decimal: indica qual separador decimal

```
# Visualização dos dados usando biblioteca que melhora a demonstração dos dados
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos por unidade da federação")
plt.show()
```
