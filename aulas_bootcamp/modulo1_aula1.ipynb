```
import pandas as pd
dados = pd.read_csv("A151346189_28_143_208.csv", encoding="ISO-8859-1",
                    skiprows=3, sep=';', skipfooter=12, thousands=".", decimal=",")
dados.head()
```

**Leitura do arquivo**
* Encoding: serve para interpretar certo acentuação
* skiprows: ignora linhas desnecessárias do ínicio do arquivo
* sep: separador do arquivo
* skipfooter: ignora linha desnecessárias do final do arquivo
* thousand: indica qual deve ser o separador de milhar
* decimal: indica qual separador decimal




```
# Serve para ser as 5 primeiras linhas do arquivo
dados.tail()
```

```
# Média dos dados
dados.mean()
```

```
# Informação sobre o dataframe
dados.info()
```

```
# Define a maneira de mostrar os dados quando float
pd.options.display.float_format = "{:.2f}".format
```

```
# Pega uma parte específica do dataframe, também conhecida como Series
dados["2008/Ago"]
```

```
# Calculo de média de apenas uma série
dados["2008/Ago"].mean()
```

```
# Visualização de dados simples, sem extender uma biblioteca para aperfeiçoamento
# x: qual dado do eixo X
# y: qual dado do eixo y
# kind: altera o tipo do gráfico
# figsize: altera o tamanho da imagem
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

```
# Visualização dos dados usando biblioteca que melhora a demonstração dos dados
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos por unidade da federação")
plt.show()
```

```
# Desafio da aula: melhorar titulo do gráfico anterior e mostrar dados do mês de agosto

dados["2020/Ago"] = pd.to_numeric(dados["2020/Ago"], errors='coerce')

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2020/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos do mês Agosto/2020 por unidade da federação")
plt.show()
```
